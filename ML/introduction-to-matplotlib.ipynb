## Making figure with numpy arrays

```python
import numpy as np
```

```python
#create some data
```

```python
x=np.linspace(0,10,100)
x
```

```python
#plot the data
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
#use the same data and make scatter plot
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
#make a plot from dictionary
nut={"Almond":10,"Peanut":6,"cashew":12}
fig,ax=plt.subplots()
ax.bar(nut.keys(),nut.values());
ax.set(title="Prices");
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut.keys()),list(nut.values()));
```

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
#Subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=
                                       (10,5));
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut.keys(),nut.values());
ax4.hist(np.random.randn(1000));
```
