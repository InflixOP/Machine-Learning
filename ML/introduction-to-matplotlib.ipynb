## Plotting from pandas dataframes

```python
import pandas as pd
```

```python
car_sales=pd.read_csv("car-sales.csv")
```

```python
car_sales
```

```python
ts=pd.Series(np.random.randn(1000),index=pd.date_range("1/1/2020",periods=1000))
ts
```

```python
ts=ts.cumsum()
ts.plot();
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','',regex=True)
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Sale Date"]=pd.date_range("1/1/2020",periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

```python
#lets plot total sales
car_sales.plot(x="Sale Date",y="Total Sales");
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter")
```

```python
#lets try on another dataset
heart=pd.read_csv("heart-disease.csv")
heart
```

```python
heart["age"].plot.hist();
```

```python
heart.head()
```

```python
heart.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50=heart[heart["age"]>50]
```

```python
over_50
```

```python
over_50.plot(kind="scatter",x='age',y='chol',c='target');
```

```python
#OO method
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x='age',y='chol',c='target',ax=ax);
#ax.set-xlim([45,100]);
```

```python
##OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
ax.set(title="Heart disease",xlabel="age",ylabel="chol")
ax.legend(*scatter.legend_elements(),title="Target")

##Add a horizontal line

ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
#subplot of age,chol,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10));
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
ax0.set(title="heart",xlabel="age",ylabel="chol");
ax0.legend(*scatter.legend_elements(),title="Targert");
ax0.axhline(y=over_50["chol"].mean(),linestyle='--');
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"]);
ax1.set(title="heart",xlabel="age",ylabel="thalach");
ax1.legend(*scatter.legend_elements(),title="Targert");
ax1.axhline(y=over_50["thalach"].mean(),linestyle='--');
fig.suptitle("Heart diseae",fontsize=16,fontweight="bold");
```
